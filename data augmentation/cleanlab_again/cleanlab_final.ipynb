```python
import pandas as pd

origin = pd.read_csv('filtered_final_triple.csv')
first = pd.read_csv('filtered_final_triple_first_cleanlab.csv')
second = pd.read_csv('filtered_final_triple_second_cleanlab.csv')
third = pd.read_csv('filtered_final_triple_third_cleanlab.csv')
```

```python
origin = origin.drop(columns=['Unnamed: 0'])
first = first.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
second = second.drop(columns=['Unnamed: 0.2', 'Unnamed: 0.1', 'Unnamed: 0'])
third = third.drop(columns=['Unnamed: 0.3', 'Unnamed: 0.2', 'Unnamed: 0.1', 'Unnamed: 0'])
```

```python
origin.to_csv('filtered_final_triple.csv', index=False)
first.to_csv('filtered_final_triple_first_cleanlab.csv', index=False)
second.to_csv('filtered_final_triple_second_cleanlab.csv', index=False)
third.to_csv('filtered_final_triple_third_cleanlab.csv', index=False)
```

### target name 설정

```python
target_to_domain = {
    '생활문화': 0,
    '스포츠': 1,
    '정치': 2,
    '사회': 3,
    'IT과학': 4,
    '경제': 5,
    '세계': 6
}
```

```python
domain_to_target = {v: k for k, v in target_to_domain.items()}
```

```python
origin['target_name'] = origin['target'].map(domain_to_target)
first['target_name'] = first['target'].map(domain_to_target)
second['target_name'] = second['target'].map(domain_to_target)
third['target_name'] = third['target'].map(domain_to_target)
```

### origin vs first_cleanlab

```python
common_index = origin.index.intersection(first.index)
origin_changed = origin.loc[common_index][origin.loc[common_index, 'target'] != first.loc[common_index, 'target']]
first_changed = first.loc[common_index][origin.loc[common_index, 'target'] != first.loc[common_index, 'target']]
origin_changed['first_target'] = first_changed['target_name']
origin_changed
```

```python
common_index = first.index.intersection(second.index)
first_changed = first.loc[common_index][first.loc[common_index, 'target'] != second.loc[common_index, 'target']]
second_changed = second.loc[common_index][first.loc[common_index, 'target'] != second.loc[common_index, 'target']]
first_changed['second_target'] = second_changed['target_name']
first_changed.to_csv('compare_second.csv')
```

```python
common_index = second.index.intersection(third.index)
second_changed = second.loc[common_index][second.loc[common_index, 'target'] != third.loc[common_index, 'target']]
third_changed = third.loc[common_index][second.loc[common_index, 'target'] != third.loc[common_index, 'target']]
second_changed['third_target'] = third_changed['target_name']
second_changed
```

### 아웃라이어 필터링 결과 확인

```python
all_drop = pd.read_csv('filtered_second_cleanlab_ids_outlier.csv')
little_drop = pd.read_csv('filtered_third_cleanlab_ids_outlier.csv')
```

```python
# 각 target별 개수와 비율 구하기
target_counts = all_drop['target'].value_counts()
target_ratios = all_drop['target'].value_counts(normalize=True)

# 결과 출력
result_df = pd.DataFrame({
    'Count': target_counts,
    'Ratio': target_ratios
})

print(result_df)
```
