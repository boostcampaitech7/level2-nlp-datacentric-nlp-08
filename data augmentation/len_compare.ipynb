## 글자 수 비교
- 글자 수 차이 threshold: 19
    - `filtered_df`: 글자수 차이 19 이상
    - `will_use`: 글자수 차이 19 미만

```python
import pandas as pd

ner_df = pd.read_csv('../data/train/ner_train.csv')
```

```python
ner_df['compare_len'] = abs(ner_df['text'].str.len() - ner_df['re_text'].str.len())
```

```python
ner_df = ner_df.drop(columns=['Unnamed: 0'])
```

```python
ner_df
```

```python
# threshold: 19
filtered_df = ner_df[ner_df['compare_len']>=19]
filtered_df
```

```python
will_use = ner_df[ner_df['compare_len']<19]
will_use
```

```python
will_use_ID = will_use['ID'].to_list()
```

```python
noisy_df = pd.read_csv('../train_noisy.csv')
will_use_noisy = noisy_df[noisy_df['ID'].isin(will_use_ID)]
```

```python
will_use_noisy.to_csv('will_use_noisy.csv')
```

```python
# 각 target별 개수와 비율 구하기
target_counts = will_use['target'].value_counts()
target_ratios = will_use['target'].value_counts(normalize=True)

# 결과 출력
result_df = pd.DataFrame({
    'Count': target_counts,
    'Ratio': target_ratios
})

print(result_df)
```
