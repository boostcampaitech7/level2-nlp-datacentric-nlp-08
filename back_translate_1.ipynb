```python
import pandas as pd
source_df=pd.read_csv("resources/pre_processed_data/train_noisy.csv")
source_df=source_df.rename(columns={'text':'original_text','re_text':'text'})
```

```python
source_df
```

```python
import re
def delete_punctuations(text):
    text= re.sub(r'[!"#$%&\'()*+,-./:;<=>?@\[\]^_\`{|}~\\\\]','', text)
    return text
```

```python
source_df['text']=source_df['text'].apply(delete_punctuations)
```

```python
from googletrans import Translator
from tqdm import tqdm
import time

translator=Translator()

def backtranslate(text,src='ko',dest='en'):
    while True:
        try:
            res=translator.translate(text,src=src,dest=dest)
            return res.text
        except:
            time.sleep(2)
            continue
```

```python
source_df['eng_text']=source_df['text'].apply(backtranslate)
```

```python
source_df.to_csv("test.csv",encoding='utf-8-sig')
```

```python
tqdm.pandas()
```

```python
source_df['back_trans_txt']=source_df['eng_text'].progress_apply(lambda x : backtranslate(x,src='en',dest='ko'))
```

```python
source_df.to_csv("test.csv",encoding='utf-8-sig')
```
