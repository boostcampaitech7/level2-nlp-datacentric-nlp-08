```python
label_encodeing_dic = {'생활문화': 0, '스포츠': 1, '정치': 2, '사회': 3, 'IT과학': 4, '경제': 5, '세계': 6}
label_decoding_dic = {v:k for k, v in label_encodeing_dic.items()}
```

```python
def is_noisy_text(text, threshold=0.06):
    special_char_ratio = sum(not (c.isalnum() or c.isspace()) for c in text) / len(text)
    return special_char_ratio > threshold
```

```python
lst = []
for _, row in df.iterrows():
    id = row['ID']
    
    re_text = row['re_text']
    text = re_text.split('교정:')[-1]
    text = text.split('\n')[0]
    text = text.split('[|endofturn|]')[0].strip()
    if text == '':
        text = row['text']
        target = row['target']
        target_name = label_decoding_dic[target]
        lst.append([id, text, target, target_name])
        continue
    # if is_noisy_text(text):
    gen_topic = row['gen_topic']
    target_name = gen_topic.split('토픽:')[-1].strip()
    target_name = target_name.split('\n')[0]
    target_name = target_name.split('[|endofturn|]')[0].strip()
    try:
        target = label_encodeing_dic[target_name]
    except:
        target = row['target']
        target_name = label_decoding_dic[target]
        lst.append([id, text, target, target_name])
    lst.append([id, text, target, target_name])
    # else:
    #     target = row['target']
    #     target_name = label_decoding_dic[target]
    #     lst.append([id, text, target, target_name])
```
