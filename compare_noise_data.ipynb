```python
import pandas as pd
```

# 노이즈 체크

```python
df_noise_s = pd.read_csv("resources/processed/simon/train_noise.csv")
df_noise_n = pd.read_csv("resources/processed/noa/train_noisy_selection.csv", index_col=0)


df_not_noise_s = pd.read_csv("resources/processed/simon/train_not_noise.csv")
df_not_noise_n = pd.read_csv("resources/processed/noa/train_not_noisy_selection.csv", index_col=0)

df_noise_s2 = pd.read_csv("resources/processed/simon/train_noise2.csv")
df_not_noise_s2 = pd.read_csv("resources/processed/simon/train_not_noise2.csv")
```

```python
print("사이먼 데이터 길이", len(df_noise_s), len(df_not_noise_s), len(df_noise_s) + len(df_not_noise_s))
print("노아 데이터 길이", len(df_noise_n), len(df_not_noise_n), len(df_noise_n) + len(df_not_noise_n))
```

```python
# 노이즈 겹치는 것
noise_intersection_set = set(df_noise_s['ID']) & set(df_noise_n['ID'])
```

```python

df_noise_s_inter = df_noise_s[df_noise_s['ID'].isin(noise_intersection_set)]
df_noise_s_inter = df_noise_s_inter.sort_values(by=["ID"], ascending=[True])

df_noise_n_inter = df_noise_n[df_noise_n['ID'].isin(noise_intersection_set)]
df_noise_n_inter = df_noise_n_inter.sort_values(by=["ID"], ascending=[True])
```

```python
df_noise_s_inter['text'].tolist() == df_noise_n_inter['text'].tolist()
```

```python
# 사이먼 겹치지 않는 데이터 확인
df_noise_s[~df_noise_s['ID'].isin(noise_intersection_set)]
```

```python
# 노아 겹치지 않는 데이터 확인
for x in df_noise_n[~df_noise_n['ID'].isin(noise_intersection_set)]['text']:
    print(x)
```
