## 증강 데이터 합치기 (사회(3), IT과학(4), 경제(5))

```python
import pandas as pd

social = pd.read_csv('social.csv')
economy = pd.read_csv('economy.csv')
science = pd.read_csv('science.csv')
```

```python
social = social.rename(columns={'predicted_value': 'target'})
```

```python
aug = pd.concat([social, science, economy], axis=0, ignore_index=True)
aug = aug.drop(columns=['Unnamed: 0'])
aug
```

```python
all_drop = pd.read_csv('filtered_second_cleanlab_ids_outlier.csv')
little_drop = pd.read_csv('filtered_third_cleanlab_ids_outlier.csv')
```

```python
aug_all_drop = pd.concat([all_drop, aug], axis=0, ignore_index=True)
aug_all_drop
```

```python
# 각 target별 개수와 비율 구하기
target_counts = aug_all_drop['target'].value_counts()
target_ratios = aug_all_drop['target'].value_counts(normalize=True)

# 결과 출력
result_df = pd.DataFrame({
    'Count': target_counts,
    'Ratio': target_ratios
})

print(result_df)
```

```python
aug_little_drop = pd.concat([little_drop, aug], axis=0, ignore_index=True)
aug_little_drop
```

```python
# 각 target별 개수와 비율 구하기
target_counts = aug_little_drop['target'].value_counts()
target_ratios = aug_little_drop['target'].value_counts(normalize=True)

# 결과 출력
result_df = pd.DataFrame({
    'Count': target_counts,
    'Ratio': target_ratios
})

print(result_df)
```

```python
aug_all_drop.to_csv('filtered_second_cleanlab_ids_outlier_aug.csv')
aug_little_drop.to_csv('filtered_third_cleanlab_ids_outlier_aug.csv')
```
