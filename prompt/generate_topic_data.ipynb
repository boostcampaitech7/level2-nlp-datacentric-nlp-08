```python
import pandas as pd
```

```python
df = pd.read_csv('./resources/processed/train_not_noisy_generation_topic.csv')
```

```python
df
```

```python
label_encodeing_dic = {'생활문화': 0, '스포츠': 1, '정치': 2, '사회': 3, 'IT과학': 4, '경제': 5, '세계': 6}
label_decoding_dic = {v:k for k, v in label_encodeing_dic.items()}
```

```python
lst = []
for _, row in df.iterrows():
    id = row['ID']
    # target_name = row['target_name']
    re_text = row['re_text']
    gen_topic = row['gen_topic']
    target = gen_topic.split('주제 번호:')[1].split('->')[0].split('\n')[0].strip()
    lst.append([id, re_text, int(target), label_decoding_dic[int(target)]])
```

```python
df_topic = pd.DataFrame(lst, columns = ['ID', 'text', 'target', 'target_name'])
```

```python
df_topic.to_csv('resources/processed/train_not_noisy_topic_update.csv', encoding='utf-8-sig')
```
