```python
import pandas as pd
```

```python
df = pd.read_csv('../resources/processed/train_noisy_generation_topic.csv')
```

```python
df
```

```python
topic_keyword_dic = {'생활문화': [], '스포츠': [], '정치': [], '사회': [], 'IT과학': [], '경제': [], '세계': []}
for _, row in df.iterrows():
    target_name = row['target_name']
    print(target_name)
    gen_topic = row['gen_topic']
    topic = gen_topic.split('키워드:')[-1].split('\n')[0].strip()
    topic = topic.replace('"','')
    topic_lst = topic.split(',')
    topic_lst = [x.strip() for x in topic_lst]
    print(topic_lst)
    print(len(topic_lst))
    print('=========')
    topic_keyword_dic[target_name].extend(topic_lst)
    
```

```python
lst = []
for k, v in topic_keyword_dic.items():
    print(k)
    v = list(set(v))
    v = [x for x in v if x]
    print(v)
    print(len(v))
    lst.append([k, v])
    
```

```python
pd.DataFrame(lst, columns=['topic', 'keyword']).to_csv('train_noisy_keyword.csv', encoding='utf-8-sig', index=0)
```
