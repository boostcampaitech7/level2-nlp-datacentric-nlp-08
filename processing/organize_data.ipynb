```python
import pandas as pd
```

# headline + backtrans

```python
df_noisy_headline = pd.read_csv('./resources/processed/train_noisy_headline_update.csv', index_col=0)
df_back_trans = pd.read_csv('./resources/processed/for_merge/backtranslate_ver2.csv', index_col=0)
```

```python
df_noisy_headline
```

```python
df_back_trans
```

```python
lst = []
for _, row in df_noisy_headline.iterrows():
    id = row['ID']
    target = row['target']
    retext = row['re_text'].strip()
    lst.append([id, retext, target])
```

```python
for _, row in df_back_trans.iterrows():
    id = row['ID']
    target = row['target']
    text = row['text'].strip()
    lst.append([id, text, target])
```

```python
df_noisy_aug = pd.DataFrame(lst, columns = ["ID", "text", "target"])
```

```python
df_noisy_aug = df_noisy_aug.sort_values('ID')
```

```python
df_noisy_aug = df_noisy_aug.drop_duplicates()
```

```python
df_noisy_aug = df_noisy_aug.reset_index(drop=True)
```

```python
label_encodeing_dic = {'생활문화': 0, '스포츠': 1, '정치': 2, '사회': 3, 'IT과학': 4, '경제': 5, '세계': 6}
label_decoding_dic = {v:k for k, v in label_encodeing_dic.items()}
```

```python
df_noisy_aug['target_name'] = [label_decoding_dic[x] for x in df_noisy_aug['target']]
```

```python
df_noisy_aug['source'] = 'noisy'
```

```python
df_noisy_aug
```

# not_noisy 데이터 추가

```python
df_non = pd.read_csv('resources/processed/for_merge/simon_non_noise_third_train.csv', index_col=0)
```
