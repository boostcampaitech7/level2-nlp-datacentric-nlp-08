```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForMaskedLM

tokenizer = AutoTokenizer.from_pretrained("klue/bert-base")
model = AutoModelForMaskedLM.from_pretrained("klue/bert-base")
```

```python
import torch
# 北 김정은 ""자주통일 새시대 열어나가야""(2보)
fake = "北 김정은 [MASK][MASK]자주통일 새시대 열어나가야[MASK][MASK][MASK]2보[MASK]"

fake_tokens = tokenizer.tokenize(fake)
fake_inputs = tokenizer.encode(fake, return_tensors="pt")

bert_outputs = model(fake_inputs)
bert_outputs = torch.argmax(bert_outputs['logits'], axis=-1)
print(bert_outputs)
bert_outputs = tokenizer.decode(bert_outputs[0], skip_special_tokens=True)
bert_outputs
```

```python
from transformers import ElectraTokenizer, ElectraForMaskedLM
import torch

tokenizer = ElectraTokenizer.from_pretrained("google/electra-small-discriminator")
model = ElectraForMaskedLM.from_pretrained("google/electra-small-discriminator")

def clean_text(text):
    inputs = tokenizer(text, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    
    logits = outputs.logits
    predicted_token_ids = torch.argmax(logits, dim=-1)
    
    cleaned_text = tokenizer.decode(predicted_token_ids[0])
    return cleaned_text

noisy_text = "Thsi is an exmaple of nosiy text."
cleaned_text = clean_text(noisy_text)
print(cleaned_text)
```

```python
import re

def preprocess_and_mask_text(text):
    # 패턴 정의: 한글, 숫자, 한자, 공백만 남기고 나머지 제거
    pattern = re.compile(r'[^가-힣A-Z一-龯0-9 ]')
    
    # 마스크를 추가할 위치 확인 및 문자 제거
    preprocessed_text = []
    for char in text:
        if not pattern.match(char) or char == ' ':  # 유효한 문자(한글, 숫자, 한자)나 공백이면 그대로 유지
            preprocessed_text.append(char)
        else:  # 유효하지 않은 문자면
            if preprocessed_text and preprocessed_text[-1]=='[MASK]':
                continue
            preprocessed_text.append('[MASK]')

    # 최종 텍스트 결합
    final_text = ''.join(preprocessed_text)
    return final_text.strip()

# 테스트
text = "분당X'k gIdi}X'3·1 야ekK대^)t대립"
result = preprocess_and_mask_text(text)
print(result)
```

```python
import re

def preprocess_and_mask_text(text):
    # 패턴 정의: 한글, 숫자, 한자, 공백만 남기고 나머지 제거
    pattern = re.compile(r'[^가-힣A-Z一-龯0-9 ]')
    texts = text.split()
    # 마스크를 추가할 위치 확인 및 문자 제거
    preprocessed_text = []
    for text in texts:
        for char in text:
            if not pattern.match(char):
                continue
            preprocessed_text.append('[MASK]')
            break
        else:
            preprocessed_text.append(text)

    # 최종 텍스트 결합
    final_text = ' '.join(preprocessed_text)
    return final_text.strip()

# 테스트
text = "무선 장비로 침입자 막는다… KT텔레캅 새 보안 서비스종합"
result = preprocess_and_mask_text(text)
print(result)
```
