```python
import pandas as pd
data=pd.read_csv("resources/raw_data/korean_filtered.csv")
data
```

```python
import random

random.seed(10)
```

```python
def add_noise_to_text(text):
    # 공백을 제외한 글자들로만 작업하기 위해 리스트 생성
    chars = list(text)
    
    # 노이즈로 사용할 랜덤 문자 리스트
    random_chars = [']', '&', 'N', '%', ')', '/', 'F', 'I', '>', '@', '!', '#', '^', '*', '-', '+', '=', 'W', 'P', 'C', 'H', 'R']
    
    num_chars_to_change = 3  # 변경할 문자 수
    
    # 공백이 아닌 문자의 인덱스만 선택하도록 리스트 생성
    non_space_indices = [i for i, c in enumerate(chars) if c != ' ']
    for _ in range(num_chars_to_change):
        if not non_space_indices:
            break  # 만약 공백이 아닌 문자가 없다면 루프 중지
        # 공백이 아닌 문자 위치 중에서 랜덤한 위치 선택
        position = random.choice(non_space_indices)
        # 랜덤 문자를 선택하여 그 위치에 넣음
        random_char = random.choice(random_chars)
        chars[position] = random_char
        # 중복되지 않도록 선택한 인덱스 제거
        non_space_indices.remove(position)
    
    # 수정된 문자들을 다시 문자열로 변환
    noisy_text = ''.join(chars)
    return noisy_text
```

```python
data['noise_added_text'] = data['text'].apply(add_noise_to_text)
```
