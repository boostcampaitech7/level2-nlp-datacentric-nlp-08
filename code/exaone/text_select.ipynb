```python
import pandas as pd
data=pd.read_csv("resources/raw_data/train.csv")
data
```

이 정규표현식 ^[가-힣·.…\s]+$는 문자열 전체가 다음 조건을 만족하는 경우에만 매칭됩니다:

- 문자열이 반드시 한글 문자, ·, ., … 또는 공백 문자로만 이루어져 있어야 한다.
- 문자열은 시작부터 끝까지 위의 조건을 만족하는 문자가 하나 이상 이어져야 한다.

```python
import re

def extract_korean_only(text):
    pattern = r"^[가-힣·.…\s]+$"
    return bool(re.match(pattern, text))
```

```python
data['korean_only'] = data['text'].apply(extract_korean_only)
korean_only_df = data[data['korean_only']]
```

```python
korean_only_df=korean_only_df.drop(['Unnamed: 0'],axis=1).reset_index(drop=True)
korean_only_df.to_csv("resources/raw_data/korean_filtered.csv",encoding='utf-8-sig',index=False)
```
