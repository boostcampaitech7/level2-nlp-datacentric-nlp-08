```python
import pandas as pd
```

```python
df = pd.read_csv('./resources/processed/train_noisy_generation_headline.csv')
```

```python
df
```

```python
lst = []
for _, row in df.iterrows():
    id = row['ID']
    target = row['target']
    target_name = row['target_name']
    text = row['text']
    re_text = row['re_text']
    lst.append([id, text, target, target_name, re_text])
    gen_headline = row['gen_headline']
    head1 = gen_headline.split('헤드라인:')[1].split('\n')[0].strip()
    head1 = head1.replace('"','')
    lst.append([id, text, target, target_name, head1])
    head2 = gen_headline.split('헤드라인:')[2].split('\n')[0].strip().replace('"','').split('[|endofturn|]')[0]
    lst.append([id, text, target, target_name, head2])
```

```python
df_head = pd.DataFrame(lst, columns = ['ID', 'text', 'target', 'target_name', 're_text'])
```

```python
df_head
```

```python
df_head_duplicate = df_head.drop_duplicates(subset=['re_text'])
```

```python
df_head[df_head['ID'].isin(df_head_duplicate['ID'].tolist())]
```

```python
df_head.to_csv('train_noisy_headline.csv', encoding='utf-8-sig')
```
