# train_noisy_headline 생성한 데이터셋 filtering

```python
import pandas as pd
from tqdm import tqdm
tqdm.pandas()
source_df=pd.read_csv("resources/pre_processed_data/train_noisy_headline.csv")
```

```python
source_df=source_df.rename(columns={'text':'original_text','re_text':'text'})
source_df=source_df.drop(['Unnamed: 0'],axis=1)
source_df
```

**punctuations exclude(filtering)**

```python
import re

def filter_punctuations(text):
    pattern=r'[!"#$%&\'()*+,-./:;<=>?@\[\]^_\`{|}~\\\\]'
    return bool(re.match(pattern,text))
```

```python
source_df['still_noisy']=source_df['text'].progress_apply(filter_punctuations)
```

```python
print(source_df[source_df['still_noisy']])
print(len(source_df[source_df['still_noisy']]))
```

```python
filtered_df=source_df[~source_df['still_noisy']]
print(filtered_df)
print(len(filtered_df))
```

```python
filtered_df=filtered_df.drop_duplicates(subset=['text'])
```
